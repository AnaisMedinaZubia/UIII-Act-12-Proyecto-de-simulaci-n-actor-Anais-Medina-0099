```python
# --- SISTEMA DE IA: AGENTE DE CASTING "TALENT-SCAN POO" ---

# Lista global para almacenar las instancias de objetos Actor [7, 8]
base_datos_actor = []

class Actor:
    """Clase que define los atributos y comportamiento de un Actor [3, 9]"""

    def __init__(self, nombre, agencia, obras, premios, habilidades, edad, proyecto):
        # El método __init__ es el constructor que inicializa los atributos [10-12]
        # self es la referencia a la instancia actual del objeto [9, 13]
        self.nombre = nombre
        self.agencia = agencia
        self.obras = obras
        self.premios = premios
        self.habilidades = habilidades.lower()
        self.edad = edad
        self.proyecto = proyecto

    def evaluar_perfil_ia(self):
        """Simulación de IA: Analiza el perfil mediante lógica condicional [5, 14]"""
        # Se utilizan operadores lógicos y relacionales para la toma de decisiones [15, 16]
        if self.premios >= 3 and self.obras > 15:
            return "CATEGORÍA: ESTRELLA DE ÉLITE (Apto para Proyectos Globales)"
        elif "artes marciales" in self.habilidades or "acrobacia" in self.habilidades:
            return "CATEGORÍA: ESPECIALISTA EN ACCIÓN (Apto para Stunts)"
        elif self.edad < 22 and self.obras <= 3:
            return "CATEGORÍA: NUEVA PROMESA (Apto para Series Juveniles)"
        else:
            return "CATEGORÍA: ACTOR VERSÁTIL (Apto para Elenco de Soporte)"

# --- FUNCIONES DE GESTIÓN (CRUD) ---

def agregar_actor():
    print("\n--- REGISTRO DE TALENTO ---")
    nombre = input("Nombre: ")
    agencia = input("Agencia: ")
    obras = int(input("Obras realizadas: "))
    premios = int(input("Premios ganados: "))
    habilidades = input("Habilidades (ej: Canto, Baile, Esgrima): ")
    edad = int(input("Rango de edad: "))
    proyecto = input("Último proyecto: ")

    # Se crea una nueva instancia (objeto) de la clase Actor [17, 18]
    nuevo_actor = Actor(nombre, agencia, obras, premios, habilidades, edad, proyecto)
    base_datos_actor.append(nuevo_actor)
    print(f"¡{nombre} registrado con éxito!")

def listar_actor():
    if len(base_datos_actor) == 0:
        print("\n[!] No hay actores en la base de datos.")
    else:
        # Encabezado de la tabla alineado con f-strings [19, 20]
        header = f"{'NOMBRE':<15} | {'AGENCIA':<15} | {'OBRAS':<6} | {'PREMIOS':<8} | {'EDAD':<5} | {'PROYECTO':<15}"
        print("\n" + "="*85)
        print(header)
        print("-" * 85)

        # Ciclo for para recorrer la lista de objetos [21, 22]
        for a in base_datos_actor:
            print(f"{a.nombre:<15} | {a.agencia:<15} | {a.obras:<6} | {a.premios:<8} | {a.edad:<5} | {a.proyecto:<15}")
        print("="*85)

def actualizar_actor():
    nombre_buscar = input("\nNombre del actor a actualizar: ")
    encontrado = False
    for a in base_datos_actor:
        if a.nombre == nombre_buscar:
            print(f"Modificando datos de {a.nombre}...")
            a.agencia = input("Nueva Agencia: ")
            a.obras = int(input("Nuevas Obras: "))
            a.premios = int(input("Nuevos Premios: "))
            a.proyecto = input("Nuevo Proyecto: ")
            encontrado = True
            print("Datos actualizados.")
            break
    if not encontrado:
        print("Actor no encontrado.")

def borrar_actor():
    nombre_borrar = input("\nNombre del actor a eliminar: ")
    for a in base_datos_actor:
        if a.nombre == nombre_borrar:
            base_datos_actor.remove(a)
            print(f"¡{nombre_borrar} eliminado!")
            return
    print("Actor no encontrado.")

def ejecutar_ia():
    print("\n[IA] Iniciando escaneo de perfiles para Casting...")
    if len(base_datos_actor) == 0:
        print("Sin datos para analizar.")
    else:
        for a in base_datos_actor:
            # Se llama al método del objeto para obtener el diagnóstico de la IA [23, 24]
            resultado = a.evaluar_perfil_ia()
            print(f"Actor: {a.nombre} -> {resultado}")

# --- MENÚ PRINCIPAL ---
while True:
    print("\n========================================")
    print("      SISTEMA IA: TALENT-SCAN 3.0      ")
    print("========================================")
    print("1. Agregar Actor")
    print("2. Listar Actores (Tabla)")
    print("3. Actualizar Actor")
    print("4. Borrar Actor")
    print("5. IA: Análisis de Perfiles")
    print("6. Salir")

    opcion = input("\nSeleccione una opción: ")

    if opcion == "1":
        agregar_actor()
    elif opcion == "2":
        listar_actor()
    elif opcion == "3":
        actualizar_actor()
    elif opcion == "4":
        borrar_actor()
    elif opcion == "5":
        ejecutar_ia()
    elif opcion == "6":
        print("Apagando sistema... ¡Corte!")
        print("Programa realizado por Anais Medina NL = 40")
        break
    else:
        print("Opción inválida.")
```
